```python
import nltk

!pip install ebooklib beautifulsoup4

from ebooklib import epub
from bs4 import BeautifulSoup
from google.colab import files

# Cargar el archivo EPUB usando Google Colab
uploaded = files.upload()

# Leer el archivo EPUB cargado
epub_filename = list(uploaded.keys())[0]  # Obtener el nombre del archivo cargado
libro = epub.read_epub(epub_filename)

# Extraer el texto del libro
texto = ""
for item in libro.get_items():
    if isinstance(item, epub.EpubHtml):  # Comprobar si el elemento es de tipo EpubHtml
        soup = BeautifulSoup(item.get_content(), 'html.parser')
        texto += soup.get_text(separator=' ')  # Añadir separador para evitar concatenación sin espacios

# Verificar si el texto fue extraído
if texto.strip():
    print("Contenido del libro cargado:", texto[:500])  # Imprimir los primeros 500 caracteres del texto
else:
    print("No se pudo extraer el contenido del libro. Por favor, verifica el formato del archivo.")
```

```python
import nltk

# Descargar stopwords y el tokenizador de NLTK
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Definir las stopwords en español
stop_words = set(stopwords.words('spanish'))

# Tokenizar el texto y filtrar las stopwords (asumiendo que 'texto' ha sido definido previamente en otro bloque de código)
palabras = word_tokenize(texto)
palabras_filtradas = [palabra for palabra in palabras if palabra.lower() not in stop_words]

print("Primeras 20 palabras sin stopwords:", palabras_filtradas[:20])
```

```python
from nltk.stem import SnowballStemmer
import spacy

# Aplicar Stemming con SnowballStemmer
stemmer = SnowballStemmer('spanish')
palabras_stemmed = [stemmer.stem(palabra) for palabra in palabras_filtradas]
print("Palabras después del stemming:", palabras_stemmed[:20])

# Descargar e instalar el modelo de spaCy para español
!python -m spacy download es_core_news_sm

# Cargar el modelo de spaCy para español y aplicar lematización
nlp = spacy.load('es_core_news_sm')
doc = nlp(" ".join(palabras_filtradas))
lemas = [token.lemma_ for token in doc]
print("Palabras después de la lematización:", lemas[:20])
```

```python
import nltk

!pip install ebooklib beautifulsoup4 langid

from ebooklib import epub
from bs4 import BeautifulSoup
from google.colab import files
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, RegexpTokenizer
from nltk.probability import FreqDist
from nltk.stem import SnowballStemmer
from nltk import bigrams, trigrams
from collections import Counter
import spacy
import langid

# Cargar el archivo EPUB usando Google Colab
uploaded = files.upload()

# Leer el archivo EPUB cargado
epub_filename = list(uploaded.keys())[0]  # Obtener el nombre del archivo cargado
libro = epub.read_epub(epub_filename)

# Extraer el texto del libro
texto = ""
for item in libro.get_items():
    if isinstance(item, epub.EpubHtml):  # Comprobar si el elemento es de tipo EpubHtml
        soup = BeautifulSoup(item.get_content(), 'html.parser')
        texto += soup.get_text(separator=' ')  # Añadir separador para evitar concatenación sin espacios

# Verificar si el texto fue extraído
if texto.strip():
    print("Contenido del libro cargado:", texto[:500])  # Imprimir los primeros 500 caracteres del texto
else:
    print("No se pudo extraer el contenido del libro. Por favor, verifica el formato del archivo.")

# Descargar stopwords y el tokenizador de NLTK
nltk.download('stopwords')
nltk.download('punkt')

# Definir las stopwords en español y añadir palabras adicionales que no aportan al análisis
stop_words = set(stopwords.words('spanish'))
stop_words.update([',', '.', ';', ':', '``', "''", 'p', 'si', 'sino', 'bien', 'cada', 'debe', 'vez', 'mismo', 'siglo', 'puede', 'menos'])  # Añadir palabras irrelevantes

# Detectar el idioma del texto completo para asegurarse de que está en español
if langid.classify(texto)[0] != 'es':
    print("El texto no está en español. Por favor, proporciona un texto en español.")

# Tokenizar el texto usando un tokenizador que elimine signos de puntuación
tokenizer = RegexpTokenizer(r'\b\w+\b')  # Solo palabras (sin puntuación)
palabras = tokenizer.tokenize(texto)

# Filtrar las stopwords
palabras_filtradas = [palabra for palabra in palabras if palabra.lower() not in stop_words]
print("Primeras 20 palabras sin stopwords:", palabras_filtradas[:20])

# Aplicar Stemming con SnowballStemmer
stemmer = SnowballStemmer('spanish')
palabras_stemmed = [stemmer.stem(palabra) for palabra in palabras_filtradas]
print("Palabras después del stemming:", palabras_stemmed[:20])

# Descargar e instalar el modelo de spaCy para español
!python -m spacy download es_core_news_lg

# Cargar el modelo de spaCy para español y aplicar lematización
nlp = spacy.load('es_core_news_sm')
doc = nlp(" ".join(palabras_filtradas))
lemas = [token.lemma_ for token in doc]
print("Palabras después de la lematización:", lemas[:20])

# Análisis de frecuencia de palabras
frecuencia = FreqDist(palabras_filtradas)
print("10 palabras más comunes:", frecuencia.most_common(10))

# Generación de Bigramas y Trigramas
bigramas = list(bigrams(palabras_filtradas))
trigramas = list(trigrams(palabras_filtradas))
print("Bigramas más comunes:", Counter(bigramas).most_common(10))
print("Trigramas más comunes:", Counter(trigramas).most_common(10))
```
